```python
results_path = "Omicron/true_benchmark_abundances.tsv"

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

results = pd.read_csv(results_path, sep='\t', header=0)
data_lineages = list(results.keys()[1:])
print(data_lineages)

data_reference_sets = [" ".join(x.split("_")[1:]) for x in list(results.iloc[:,0])]
print(data_reference_sets)

data = np.genfromtxt(results_path, delimiter="\t",skip_header=True, usecols=range(1, 9))
data_sv = data - 12.5   

data = abs(data - 12.5)

sns.heatmap(data, annot=data_sv, fmt=".1f", linewidths=.5, cmap = "viridis", 
                xticklabels= data_lineages, 
                yticklabels= data_reference_sets, 
                cbar_kws={'label': 'Absolute error'})

plt.yticks(rotation=0) 
plt.xticks(rotation=30)
plt.savefig("Omicron/figures/heatmap_v2.pdf", bbox_inches='tight')


```
